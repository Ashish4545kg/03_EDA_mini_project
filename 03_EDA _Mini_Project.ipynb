### Q2. Max, min , mean of asked amount,asked equity, asked valuation? 

```python
#1.pitcher_ask Amount max min mean:
#max
print('pitcher_ask Amount max min mean:')
ask_amt=data['pitcher_ask_amount']

a=max(ask_amt)
print('max' , ':' , a)

#min
b=min(ask_amt)
print('min' , ':', b)

#mean
c=np.mean(ask_amt)
print('mean' , ':', c)
print('\n')

#2 ask_equity min max mean
#max 
print('ask_equity min max mean')

ask_eqt=data['ask_equity']

d=max(ask_eqt)
print('max' , ':' , d)

e=min(ask_eqt)
print('min',':',e)
f=np.mean(ask_eqt)
print('mean', ':' , f)
print('\n')

#3 ask_valuation max min mean
#max
print('ask_valuation max min mean')

ask_val=data['ask_valuation']
g=max(ask_val)
print('max' , ':' , g)

#min
h=min(ask_val)
print('min' ,':' , h)

i=np.mean(ask_val)
print('mean', ':' , i)

```
